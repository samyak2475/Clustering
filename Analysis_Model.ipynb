```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"],dayfirst = True)
```

```python
df.info()
```

```python
df["Age"] = 2025 - df["Year_Birth"]
```

```python
df["Age"]
```

```python
df.head()
```

```python
df["Total_Childen"] = df["Kidhome"] + df["Teenhome"]
```

```python
df["Total_Childen"]
```

```python
df.columns
```

```python
spend_cols = ["MntWines", "MntFruits",
       "MntMeatProducts", "MntFishProducts", "MntSweetProducts",
       "MntGoldProds"]
```

```python
df["Total_Spending"] = df[spend_cols].sum(axis = 1)
```

```python
df["Total_Spending"]
```

```python
df["Customer_Since"] = (pd.Timestamp("today") - df["Dt_Customer"]).dt.days
```

```python
df["Customer_Since"]
```

```python
sns.histplot(df["Age"], bins = 30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df["Income"], bins = 30, kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
sns.histplot(df["Total_Spending"], bins = 30, kde=True)
plt.title("Total Spending Distribution")
plt.show()
```

```python
sns.boxplot(x = "Education", y = "Income", data = df)
plt.xticks(rotation = 45)
plt.title("Income by Education Level")
plt.show()
```

```python
sns.boxplot(x = "Marital_Status", y = "Total_Spending", data = df)
plt.xticks(rotation = 45)
plt.title("Total Spending by Marital Status")
plt.show()
```

```python
corr = df[["Income","Age","Recency","Total_Spending","NumWebPurchases","NumStorePurchases"]].corr()
```

```python
corr
```

```python
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation matrix")
plt.show()
```

```python
pivot_income = df.pivot_table(values = "Income", index = "Education", columns = "Marital_Status", aggfunc = "mean")
```

```python
pivot_income
```

```python
sns.heatmap(pivot_income, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Average Income by Education and Marital Status")
plt.show()
```

```python
group1 = df.groupby("Education")["Total_Spending"].mean().sort_values(ascending=False)
```

```python
group1
```

```python
group1.plot(kind = "bar", color = "skyblue")
plt.title("Average Total Spending by Education")
plt.ylabel("Average Total Spending")
plt.xticks(rotation = 45)
plt.show()
```

```python
df["AcceptedAny"] = df[["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5","Response"]].sum(axis=1)
```

```python
df["AcceptedAny"].unique()
```

```python
df["AcceptedAny"] = df["AcceptedAny"].apply(lambda x: 1 if x > 0 else 0)
```

```python
df["AcceptedAny"].unique()
```

```python
group2 = df.groupby("Marital_Status")["AcceptedAny"].mean().sort_values(ascending=False)
```

```python
df.columns
```

```python
group2
```

```python
df.head()
```

```python
group2.plot(kind = "bar", color = "orange")
plt.title("Acceptance Rate by Marital Status")
plt.ylabel("Acceptance Rate")
plt.xticks(rotation = 45)
plt.show()
```
